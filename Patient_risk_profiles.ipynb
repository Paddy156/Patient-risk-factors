```python
#Calculate correlational values
correlation = patient.corr()
#The data comparison chosen for this project is in the doesn't need the last 30 rows so singly out
data_needed = correlation.tail(n=17)
#Use data needed to remove last 30 rows and show correlation between rest
data_needed = data_needed.iloc[:,:-17]
```

```python
data_needed
```

```python
#Iterate through the rows to find the column with the highest value in each row
for i in range(len(data_needed)):
    #Get the name of the column with the highest value in the row
    prevalence = data_needed.iloc[i].idxmax()
    print(f'Column with the highest value in row {data_needed.index[i]}: {prevalence}')
```
